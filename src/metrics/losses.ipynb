```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
losses_amsterdam = pd.read_csv('../results/final_losses/losses_GRU_at_amsterdam_20250304-152113.csv', sep=';')
losses_rotterdam = pd.read_csv('../results/final_losses/losses_GRU_at_rotterdam_20250304-151850.csv', sep=';')
losses_utrecht = pd.read_csv('../results/final_losses/losses_GRU_at_utrecht_20250304-152220.csv', sep=';')
```

```python
import matplotlib.pyplot as plt

# Create figure with 3 subplots vertically arranged
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15))
fig.suptitle('Training and Validation Losses Across Cities', fontsize=16)

# Amsterdam plot
ax1.plot(losses_amsterdam['L_train'], label='Training loss', color='blue')
ax1.plot(losses_amsterdam['L_val'], label='Validation loss', color='red')
ax1.set_title('Amsterdam')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend()
ax1.grid(True)

# Rotterdam plot
ax2.plot(losses_rotterdam['L_train'], label='Training loss', color='blue')
ax2.plot(losses_rotterdam['L_val'], label='Validation loss', color='red')
ax2.set_title('Rotterdam')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

# Utrecht plot
ax3.plot(losses_utrecht['L_train'], label='Training loss', color='blue')
ax3.plot(losses_utrecht['L_val'], label='Validation loss', color='red')
ax3.set_title('Utrecht')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Loss')
ax3.legend()
ax3.grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
